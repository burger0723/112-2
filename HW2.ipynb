```python
import json

# 讀取 JSON 檔案
with open('voice.json','r',encoding='utf-8') as file:
    data = json.load(file)

# 提取 records 列表
records = data['records']

# 遍歷每個記錄
for record in records:
    city = record['city']
    date1 = record['date1']
    case1 = record['case1']
    average1 = record['average1']
    date2 = record['date2']
    case2 = record['case2']
    average2 = record['average2']
    date3 = record['date3']
    case3 = record['case3']
    average3 = record['average3']
    date4 = record['date4']
    case4 = record['case4']
    average4 = record['average4']
    
    # 輸出結果
    print(f"城市：{city}")
    print(f"日期1：{date1}，案例1：{case1}，平均1：{average1}")
    print(f"日期2：{date2}，案例2：{case2}，平均2：{average2}")
    print(f"日期3：{date3}，案例3：{case3}，平均3：{average3}")
    print(f"日期4：{date4}，案例4：{case4}，平均4：{average4}")
    print("-------------------------")
```

```python
import json
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial Unicode MS'

# 讀取JSON文件
with open('voice.json', 'r',encoding='utf-8' ) as json_file:
    data = json.load(json_file)

# 提取城市名稱和105年的案例數
cities = [record["city"] for record in data["records"][1:]]
cases = []
for record in data["records"][1:]:
    case_value = record["case1"]
    if case_value != "-":
        cases.append(int(case_value.replace(",", "")))
    else:
        cases.append(0)  # 將缺失值或無效值設置為0
        
# 繪製柱狀圖
plt.bar(cities, cases)

# 添加標籤和標題
plt.xlabel("城市")
plt.ylabel("案例數")
plt.title("104年案例數長條圖")

plt.xticks(rotation=45, ha='right')

# 顯示圖形
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial Unicode MS'

# 讀取JSON文件
with open('voice.json', 'r' ,encoding='utf-8') as json_file:
    data = json.load(json_file)

# 提取台北市的案例數
taipei_cases = []
for record in data["records"]:
    if record["city"] == "臺北市":
        for year in ["1", "2", "3", "4"]:
            case_value = record["case" + year]
            if case_value != "-":
                taipei_cases.append(int(case_value.replace(",", "")))
            else:
                taipei_cases.append(0)
        break

# 提取年度
years = ["104年度", "105年度", "106年度", "107年度"]

# 繪製區域圖
plt.fill_between(years, taipei_cases, color='blue', alpha=0.3)

# 添加標籤和標題
plt.xlabel("年度")
plt.ylabel("案例數")
plt.title("臺北市104年至107年案例數區域圖")

# 顯示圖形
plt.show()
```
