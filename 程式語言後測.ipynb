```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#讀入csv
df = pd.read_csv('./adult.csv')
      
plt.hist2d(df['Age'],df['Income'],bins=30,edgecolor='Black')
plt.title('Income data')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```
